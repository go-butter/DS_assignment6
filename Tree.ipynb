## 01-1) 리프 노드의 수

리프 노드는 자식이 없는 노드이다.

주어진 트리에서 리프 노드: **7, 17, 70**

∴ 총 **3개**

## 01-2) 높이 4를 유지하면서 리프 노드의 최대 개수

높이 $h$인 이진 트리에서 레벨 $h$에 배치 가능한 최대 노드 수는 $2^h$개이다.

현재 트리의 레벨별 구조는 다음과 같다.

| 레벨 | 노드 |
|------|------|
| 1 | 40 |
| 2 | 10, 40 |
| 3 | 7, 20, 70 |
| 4 | 17 |

높이 4를 유지하면서 리프 노드의 수를 최대화하려면, 레벨 3의 모든 노드에 자식을 최대한 추가하고, 레벨 2의 빈 자식 슬롯도 활용해야 한다.

- 레벨 2의 오른쪽 노드(40)는 현재 자식이 70 하나뿐이므로, 왼쪽 자식 1개를 레벨 3에 추가할 수 있다. → 레벨 3 노드 1개 추가
- 레벨 3의 노드는 7, 20, 70, (추가된 1개) 총 4개이며, 각각 자식을 2개씩 가질 수 있으므로 레벨 4에 최대 $4 \times 2 = 8$개 배치 가능하다.

이때 모든 리프 노드는 레벨 4에 위치하게 된다.

∴ 높이 4를 유지할 때 리프 노드의 최대 개수: **8개**

## 01-3) 높이 4 포화 이진 트리에서 리프 노드와 내부 노드의 차이

높이 $h$인 포화 이진 트리(perfect binary tree)의 전체 노드 수: $2^h - 1$

$h = 4$일 때:

- 전체 노드 수 $= 2^4 - 1 = 15$개
- 리프 노드 수 $L = 2^{h-1} = 2^3 = 8$개
- 내부 노드 수 $I = 15 - 8 = 7$개
- 차이 $= L - I = 8 - 7 =$ **1개**

## 01-4) 높이 h인 모든 포화 이진 트리에 대해 성립하는가?

높이 $h$인 포화 이진 트리에서:

$$L = 2^{h-1}, \quad I = 2^h - 1 - 2^{h-1} = 2^{h-1} - 1$$

$$L - I = 2^{h-1} - (2^{h-1} - 1) = 1$$

모든 $h$에 대해 $L - I = 1$이므로 **성립한다.**

## 02-1) 완전 이진 트리에서 총 노드 수 n일 때 리프 노드의 수

완전 이진 트리(complete binary tree)에서 총 노드 수가 $n$이면:

$$\text{리프 노드 수} = \lceil n/2 \rceil$$

- $n$이 홀수: $\frac{n+1}{2}$개
- $n$이 짝수: $\frac{n}{2}$개

## 02-2) 리프 노드와 내부 노드의 관계

내부 노드 수 $I = n - \lceil n/2 \rceil = \lfloor n/2 \rfloor$

- $n$이 홀수: $L = \frac{n+1}{2},\ I = \frac{n-1}{2}$ → $L = I + 1$
- $n$이 짝수: $L = \frac{n}{2},\ I = \frac{n}{2}$ → $L = I$

∴ 리프 노드의 수는 내부 노드의 수와 같거나 하나 더 많다: **$L = I$ 또는 $L = I + 1$**

## 03-1) 전위 순회 (Preorder: 루트 → 왼쪽 → 오른쪽)

$$5 \to 3 \to 3 \to 8 \to 7 \to 6 \to 9$$

## 03-2) 중위 순회 (Inorder: 왼쪽 → 루트 → 오른쪽)

$$3 \to 3 \to 5 \to 6 \to 7 \to 8 \to 9$$

## 03-3) 후위 순회 (Postorder: 왼쪽 → 오른쪽 → 루트)

$$3 \to 3 \to 6 \to 7 \to 9 \to 8 \to 5$$

## 03-4) 중위 순회가 항상 정렬된 순서로 방문함을 증명

**기저 사례:** $h = 1$이면 트리는 단 하나의 노드(루트)만 갖는다. 이 경우 중위 순회는 해당 노드 하나만 방문하므로 정렬된 순서가 성립한다.

**귀납 단계:** 임의의 BST $T$가 루트 $r$, 왼쪽 서브트리 $T_L$, 오른쪽 서브트리 $T_R$로 구성된다고 하자.

BST의 정의에 의해:

$$\forall k \in T_L,\ k < r \quad \text{and} \quad \forall k \in T_R,\ k > r$$

귀납 가정: $T_L$과 $T_R$ 각각에 대해 중위 순회는 정렬된 순서로 방문한다.

$T$의 중위 순회는 $T_L$ 중위 순회 $\to r \to$ $T_R$ 중위 순회 순서로 실행된다.

- $T_L$ 중위 순회의 모든 값 $< r$ (BST 성질)
- $T_R$ 중위 순회의 모든 값 $> r$ (BST 성질)
- $T_L$, $T_R$ 내부는 각각 귀납 가정에 의해 정렬됨

따라서 세 부분이 오름차순으로 연결되므로 $T$ 전체의 중위 순회도 정렬된 순서가 된다. $\blacksquare$

## 04) 키 30 검색 과정

주어진 트리(루트 55)에서 30을 검색하는 과정:

1. 루트 **55** 방문 → $30 < 55$ → **왼쪽** 이동
2. **15** 방문 → $30 > 15$ → **오른쪽** 이동
3. **28** 방문 → $30 > 28$ → **오른쪽** 이동
4. **30** 방문 → $30 = 30$ → **검색 성공**

방문 순서: $55 \to 15 \to 28 \to 30$

## 05) 키 16 삽입 후 트리 모양

삽입 경로: $55 \to 15 \to 28 \to 18$

$16 < 18$이므로 **18의 왼쪽 자식**으로 삽입된다.

```
        55
       /  ＼
      15   60
     /  ＼   ＼
    8   28   90
       / ＼
      18  30
     /      ＼
    16       48
```

## 06) 키 38 삭제 후 트리 모양

38은 자식이 하나(왼쪽 자식 33)인 노드이다.

삭제 규칙: 자식이 하나인 노드를 삭제하면 그 자식이 삭제된 노드의 자리를 대체한다.

→ 48의 왼쪽 자식이 38에서 33으로 교체된다. 33의 자식(32, 36)은 그대로 유지된다.

```
        55
       /  ＼
      15   60
     /  ＼    ＼
    8   28    90
   /       ＼
  3        48
           /  ＼
          33  50
         /  ＼
        32  36
```

## 07) 키 3 삭제 후 트리 모양

3은 리프 노드(자식 없음)이다.

삭제 규칙: 리프 노드는 그냥 제거한다.

→ 8의 왼쪽 자식 포인터가 null이 된다. 나머지 구조는 원본 그대로 유지된다.

```
        55
       /  ＼
      15   60
     / ＼    ＼
    8   28    90
          ＼
           48
           / ＼
          38  50
         /
        33
       / ＼
      32  36
```

## 08) 키 15 삭제 후 트리 모양

15는 자식이 둘(왼쪽: 8, 오른쪽: 28)인 노드이다.

삭제 규칙: 자식이 둘인 경우 중위 후속자로 대체한다.

15의 중위 후속자 = 오른쪽 서브트리에서 가장 작은 값 = 18

처리 순서:
1. 15 자리에 18을 복사한다.
2. 18의 원래 위치(28의 왼쪽 자식)에서 18을 제거한다. 18의 오른쪽 자식 30이 28의 왼쪽 자식으로 올라온다.

```
        55
       /  ＼
      28   60
     / ＼    ＼
    8   48   90
   /   / ＼
  3   38  50
     /
    33
   / ＼
  32  36
```

# 09) n개 원소 BST의 최대 높이와 최소 높이

최대 높이: 모든 노드가 한 방향으로만 연결된 편향 이진 트리일 때 발생한다.

$$h_{\max} = n$$

최소 높이: 가능한 한 균형 잡힌 완전 이진 트리에 가까운 형태일 때 발생한다.

높이 $h$인 완전 이진 트리는 최대 $2^h - 1$개의 노드를 수용하므로:

$$2^h - 1 \geq n \implies h \geq \log_2(n+1)$$

$$h_{\min} = \lceil \log_2(n+1) \rceil$$

**정리:**

| | 높이 | 트리 형태 |
|---|---|---|
| 최대 높이 | $n$ | 편향 이진 트리 |
| 최소 높이 | $\lceil \log_2(n+1) \rceil$ | 균형 이진 트리 |
