리프 노드는 자식이 없는 노드이다.

주어진 트리에서 리프 노드: **7, 17, 70**

∴ 총 **3개**

## 01-2) 높이 4를 유지하면서 추가 가능한 리프 노드의 최대 개수

높이 $h$인 이진 트리에서 레벨 $h$에 배치 가능한 최대 노드 수는 $2^h$개이다.

높이 4를 유지하면서 노드를 추가하면 기존 리프(7, 17, 70)는 내부 노드로 바뀌고, 레벨 4에 최대 $2^4 = 16$개의 노드를 배치할 수 있다.

∴ 가능한 리프 노드의 최대 개수: **16개**

---

## 01-3) 높이 4 포화 이진 트리에서 리프 노드와 내부 노드의 차이

높이 $h$인 포화 이진 트리(perfect binary tree)의 전체 노드 수: $2^{h+1} - 1$

$h = 4$일 때:

- 전체 노드 수 $= 2^5 - 1 = 31$개
- 리프 노드 수 $L = 2^4 = 16$개
- 내부 노드 수 $I = 31 - 16 = 15$개
- 차이 $= L - I = 16 - 15 =$ **1개**

---

## 01-4) 높이 h인 모든 포화 이진 트리에 대해 성립하는가?

높이 $h$인 포화 이진 트리에서:

$$L = 2^h, \quad I = 2^{h+1} - 1 - 2^h = 2^h - 1$$

$$L - I = 2^h - (2^h - 1) = 1$$

모든 $h$에 대해 $L - I = 1$이므로 **성립한다.**

## 02-1) 완전 이진 트리에서 총 노드 수 n일 때 리프 노드의 수

완전 이진 트리(complete binary tree)에서 총 노드 수가 $n$이면:

$$\text{리프 노드 수} = \lceil n/2 \rceil$$

- $n$이 홀수: $\frac{n+1}{2}$개
- $n$이 짝수: $\frac{n}{2}$개

---

## 02-2) 리프 노드와 내부 노드의 관계

내부 노드 수 $I = n - \lceil n/2 \rceil = \lfloor n/2 \rfloor$

- $n$이 홀수: $L = \frac{n+1}{2},\ I = \frac{n-1}{2}$ → $L = I + 1$
- $n$이 짝수: $L = \frac{n}{2},\ I = \frac{n}{2}$ → $L = I$

∴ 리프 노드의 수는 내부 노드의 수와 같거나 하나 더 많다: **$L = I$ 또는 $L = I + 1$**

## 03-4) 중위 순회가 항상 정렬된 순서로 방문함을 증명

**구조적 귀납법(structural induction)으로 증명**

**기저 사례:** 트리가 공백(empty)이면 방문 순서가 없으므로 정렬된 순서가 성립한다.

**귀납 단계:** 임의의 BST $T$가 루트 $r$, 왼쪽 서브트리 $T_L$, 오른쪽 서브트리 $T_R$로 구성된다고 하자.

BST의 정의에 의해:

$$\forall k \in T_L,\ k < r \quad \text{and} \quad \forall k \in T_R,\ k > r$$

귀납 가정: $T_L$과 $T_R$ 각각에 대해 중위 순회는 정렬된 순서로 방문한다.

$T$의 중위 순회는 "$T_L$ 중위 순회 $\to r \to$ $T_R$ 중위 순회" 순서로 실행된다.

- $T_L$ 중위 순회의 모든 값 $< r$ (BST 성질)
- $T_R$ 중위 순회의 모든 값 $> r$ (BST 성질)
- $T_L$, $T_R$ 내부는 각각 귀납 가정에 의해 정렬됨

따라서 세 부분이 오름차순으로 연결되므로 $T$ 전체의 중위 순회도 정렬된 순서가 된다. $\blacksquare$

주어진 트리(루트 55)에서 30을 검색하는 과정:

1. 루트 **55** 방문 → $30 < 55$ → **왼쪽** 이동
2. **15** 방문 → $30 > 15$ → **오른쪽** 이동
3. **28** 방문 → $30 > 28$ → **오른쪽** 이동
4. **30** 방문 → $30 = 30$ → **검색 성공**

방문 순서: $55 \to 15 \to 28 \to 30$

## 08) 키 15 삭제 후 트리 모양

15는 자식이 둘(왼쪽: 8, 오른쪽: 28)인 노드이다.

```
        55
       /  ＼
      18   60
     / ＼    ＼
    8   28    90
   /   / ＼
  3  30   48
          / ＼
         38  50
        /
       33
      / ＼
     32  36
```

## 09) n개 원소 BST의 최대 높이와 최소 높이

최대 높이: 모든 노드가 한 방향으로만 연결된 편향 이진 트리일 때 발생한다.

$$h_{\max} = n - 1$$

최소 높이: 가능한 한 균형 잡힌 완전 이진 트리에 가까운 형태일 때 발생한다.

높이 $h$인 완전 이진 트리는 최대 $2^{h+1} - 1$개의 노드를 수용하므로:

$$2^{h+1} - 1 \geq n \implies h \geq \log_2(n+1) - 1$$

$$h_{\min} = \lceil \log_2(n+1) \rceil - 1 = \lfloor \log_2 n \rfloor$$

**정리:**

| | 높이 | 트리 형태 |
|---|---|---|
| 최대 높이 | $n - 1$ | 편향 이진 트리 |
| 최소 높이 | $\lfloor \log_2 n \rfloor$ | 균형 이진 트리 |
